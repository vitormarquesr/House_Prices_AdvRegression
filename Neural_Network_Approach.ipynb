```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

# Preprocessing

```python
df = pd.read_csv('train.csv')
```

```python
X = df.drop(['SalePrice'], axis=1)
y = df['SalePrice']
```

```python
X_num = X[X.columns[X.dtypes != 'object']].drop(['Id'],axis=1)
```

```python
X_num = SimpleImputer(strategy='constant', fill_value=0).fit_transform(X_num)
```

```python
X_num = StandardScaler().fit_transform(X_num)
y_train = y
```

# Building Network

```python
def build_network_regressor(input_shape=[36],n_hid=2, n_neuro_hid=30, activation_hid='relu',
                 lr=1e-3, loss='mse'):
    model = keras.models.Sequential()
    model.add(keras.layers.InputLayer(input_shape=input_shape))
    for c in range(n_hid):
        model.add(keras.layers.Dense(n_neuro_hid, activation=activation_hid))
    model.add(keras.layers.Dense(1))
    optm = keras.optimizers.SGD(learning_rate=lr)
    model.compile(optimizer=optm,loss=loss)
    return model
```

```python
keras.backend.clear_session()
```

```python
model = build_network_regressor(n_hid=2, n_neuro_hid=30, activation_hid='relu', lr=1e-2,
                               loss=keras.losses.mean_squared_logarithmic_error)
early_callback = keras.callbacks.EarlyStopping(patience=20, restore_best_weights=True)
model.fit(X_num, y_train, validation_split=0.15,epochs=1000, 
          callbacks=[early_callback])
```

# Searching for the optimal Learning Rate

```python
class OptimalLearningRate(keras.callbacks.Callback):
    def __init__(self, n_iters = 100, l0=1e-5, lf=1000):
        self.learning = []
        self.loss = []
        self.accuracy = []
        self.iter = 0
        self.n_iters = n_iters
        self.l0 = l0
        self.lf = lf
    def on_batch_end(self, batch,logs=None):
        self.loss.append(logs['loss'])
        self.accuracy.append(logs['loss'])
        curr_rate = keras.backend.get_value(self.model.optimizer.learning_rate)
        self.learning.append(curr_rate)
        self.iter += 1
        rate = np.exp(np.log(self.lf/self.l0)/self.n_iters)
        keras.backend.set_value(self.model.optimizer.learning_rate, curr_rate*rate)
        if self.iter == self.n_iters:
            self.model.stop_training = True     
```

```python
keras.backend.clear_session()
model2 = build_network_regressor(n_hid=3, n_neuro_hid=300, activation_hid='relu', lr=1e-5,
                               loss=keras.losses.mean_squared_logarithmic_error)
lr_callback = OptimalLearningRate(n_iters=5000)
history = model2.fit(X_num, y_train, validation_split=0.15,
         epochs=500, callbacks=[lr_callback])
```

```python
plt.xscale('log')
plt.ylim((0, 0.4))
plt.xlim((100,1000))
plt.plot(lr_callback.learning, lr_callback.loss)
```

```python
idx = np.argmin(lr_callback.loss)
lr_change = lr_callback.learning[idx]
ideal_lr = lr_change/10
ideal_lr
```

# Training model with ideal learning rate

```python
model = build_network_regressor(n_hid=3, n_neuro_hid=300, activation_hid='relu', lr=ideal_lr,
                               loss=keras.losses.mean_squared_logarithmic_error)
early_callback = keras.callbacks.EarlyStopping(patience=20, restore_best_weights=True)
model.fit(X_num, y_train, validation_split=0.15,epochs=1000, 
          callbacks=[early_callback])
```

# Predictions for test data

```python
df_t = pd.read_csv('test.csv')
X_t = df_t[df_t.columns[X.dtypes != 'object']].drop(['Id'],axis=1)
X_t = SimpleImputer(strategy='constant', fill_value=0).fit_transform(X_t)
X_t = StandardScaler().fit_transform(X_t)

df_pred = pd.concat([df_t.Id, pd.Series(model.predict(X_t).ravel(), name='SalePrice')], axis=1)
df_pred.to_csv('First_Try_Neural_Nets.csv', index=False)
```
